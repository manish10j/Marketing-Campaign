```python
bank['converted'] = bank['y'].apply(lambda c:1 if c=='yes' else 0)
```

```python
# Drop Y column from dataframe 
bank=bank.drop('y',axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_corr(df,size=10):
    corr=df.corr()
    fig,ax= plt.subplots(figsize=(size, size))
    ax.legend()
    cax = ax.matshow(corr)
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)), corr.columns,rotation='vertical')
    plt.yticks(range(len(corr.columns)), corr.columns)
    
plot_corr(bank)
```

```python
print(f"Total Converted : {bank.converted.sum()} out of {len(bank)}")
```

```python
print(f"Total Conversion Rate: {bank.converted.sum()/len(bank)*100 :.2f}%")
```

```python
# Getting Sum of number of Conversion per campaign
```

```python
pd.DataFrame(bank.groupby('campaign')['converted'].sum())
```

```python
pd.DataFrame(bank.groupby('campaign')['converted'].count())
```

```python
# Finding Conversion rate by each campaign
```

```python
conv_by_camp=bank.groupby('campaign')['converted'].sum()/bank.groupby('campaign')['converted'].count() * 100
```

```python
conv_by_camp.head(3)
```

```python
# Alternate method using Aggregate and lambda function
```

```python
converted_sum_count=bank.groupby(['campaign','converted']).agg({'converted':'count'})
```

```python
converted_sum_count.head(3)
```

```python
conv_by_camp2=converted_sum_count.groupby(level=0).apply(lambda x : 100*x/float(x.sum()))
```

```python
conv_by_camp2
```

```python
# Conversion Rate by Age
```

```python
converted_sum_age=bank.groupby(['age','converted']).agg({'converted':'count'})
```

```python
converted_sum_age.head(3)
```

```python
conv_by_age = converted_sum_age.groupby(level=0).apply(lambda x: 100*x/float(x.sum()))
```

```python
conv_by_age
```

```python
conv_by_age=bank.groupby('age')['converted'].sum()/bank.groupby('age')['converted'].count()*100
```

```python
conv_by_age.head(3)
```

```python
conv_by_age=pd.DataFrame(conv_by_age).reset_index()
```

```python
bank['converted'] = bank['y'].apply(lambda c:1 if c=='yes' else 0)
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_corr(df,size=10):
    corr=df.corr()
    fig,ax= plt.subplots(figsize=(size, size))
    ax.legend()
    cax = ax.matshow(corr)
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)), corr.columns,rotation='vertical')
    plt.yticks(range(len(corr.columns)), corr.columns)
    
plot_corr(bank)
```

```python
# Getting Sum of number of Conversion per campaign
```

```python
# Finding Conversion rate by each campaign
```

```python
conv_by_camp=bank.groupby('campaign')['converted'].sum()/bank.groupby('campaign')['converted'].count() * 100
```

```python
conv_by_camp.head(3)
```

```python
# Alternate method using Aggregate and lambda function
```

```python
converted_sum_count=bank.groupby(['campaign','converted']).agg({'converted':'count'})
```

```python
converted_sum_count.head(3)
```

```python
conv_by_camp2=converted_sum_count.groupby(level=0).apply(lambda x : 100*x/float(x.sum()))
```

```python
conv_by_camp2
```

```python
# Conversion Rate by Age
```

```python
converted_sum_age=bank.groupby(['age','converted']).agg({'converted':'count'})
```

```python
converted_sum_age.head(3)
```

```python
conv_by_age = converted_sum_age.groupby(level=0).apply(lambda x: 100*x/float(x.sum()))
```

```python
conv_by_age
```

```python
conv_by_age=bank.groupby('age')['converted'].sum()/bank.groupby('age')['converted'].count()*100
```

```python
conv_by_age.head(3)
```

```python
conv_by_age=pd.DataFrame(conv_by_age).reset_index()
```

```python
conv_by_age.head(3)
```

```python
conv_by_age.sort_values('converted',ascending=False)
```

```python
conv_by_age.set_index('age',inplace=True)
```

```python
import matplotlib.pyplot as plt
ax = conv_by_age.plot(grid=True, figsize=(10, 7), title='Conversion Rates by Age')

ax.set_xlabel('Age')
ax.set_ylabel('Conversion Rate (%)')

plt.show()
```

```python
 # Creating Age Group for better interpretation
```

```python
bank['age_group'] = bank['age'].apply(lambda x: '[17, 25)' if x < 25 else '[25, 35)' 
                                  if x < 35 else '[35, 45)'
                                  if x < 45 else '[45, 55)' 
                                  if x < 55 else '[55, 65)'
                                  if x < 65 else '65+')
bank.head()
```

```python
pd.DataFrame(bank.groupby('age_group')['converted'].sum())
```

```python
pd.DataFrame(bank.groupby('age_group')['converted'].count())
```

```python
conv_by_age_group = bank.groupby('age_group')['converted'].sum() / bank.groupby('age_group')['converted'].count() * 100.0
```

```python
conv_by_age_group.head()
```

```python
ax = conv_by_age_group.loc[['[17, 25)', '[25, 35)', '[35, 45)', '[45, 55)', '[55, 65)', '65+']].plot(
    kind='bar',
    color='coral',
    grid=True,
    figsize=(12, 8),
    title='Conversion Rates by Age Groups')

ax.set_xlabel('age')
ax.set_ylabel('conversion rate (%)')

plt.show()
```

```python
conversions_by_education = pd.pivot_table(bank,values='age',index='education', columns='converted', aggfunc=len)
conversions_by_education
```

```python
conversions_by_education.columns = ['non_conversions', 'conversions']
conversions_by_education
```

```python
# Ploting those on a piechart
```

```python
conversions_by_education.plot(
    kind='pie',
    figsize=(15, 7),
    startangle=90,
    subplots=True,
    autopct=lambda x: '%0.2f%%' % x,
    legend=False
)

plt.show()
```

```python
# Printing the column names so we can use them as labels in our next chart
list(conversions_by_education.reset_index()['education'])
```

```python
new_order = [1,0]
conversions_by_education = conversions_by_education[conversions_by_education.columns[new_order]]
conversions_by_education
```

```python
#Plot stacked bargraph
```

```python
ax = conversions_by_education.loc[['basic.4y','basic.6y','basic.9y','high.school','illiterate','professional.course',
 'university.degree','unknown']].plot(
    kind='bar', 
    grid=True,
    stacked=True,
    figsize=(10,7))

ax.set_title('Conversions by Education Level')
ax.set_xlabel('Education Level')
ax.set_ylabel('conversion counts')

plt.show()
```

```python
conversions_by_education['per_converted'] = conversions_by_education['conversions'] / conversions_by_education['non_conversions']
conversions_by_education
```

```python
ax = conversions_by_education.loc[['basic.4y','basic.6y','basic.9y','high.school','illiterate','professional.course',
 'university.degree','unknown']].plot(
    y = 'per_converted',
    kind='bar', 
    grid=True,
    stacked=True,
    figsize=(12,7))

ax.set_title('Conversions by Education Level')
ax.set_xlabel('Education Level')
ax.set_ylabel('conversion counts')

plt.show()
```

```python
# Let's explore Conversion Rate by Job
```

```python
conv_rate_by_job = bank.groupby('job')['converted'].sum() / bank.groupby('job')['converted'].count() * 100.0
conv_rate_by_job
```

```python
# Sorting it for a better aesthetic when plotted
conv_rate_by_job = conv_rate_by_job.sort_values()
conv_rate_by_job
```

```python
ax = conv_rate_by_job.plot(
    kind='barh',
    color='coral',
    grid=True,
    figsize=(10, 8),
    title='Conversion Rates by Job')

ax.set_xlabel('conversion rate (%)')
ax.set_ylabel('Job')

plt.show()
```
